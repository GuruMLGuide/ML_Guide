```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()

# Load the dataset
data = pd.read_csv('house price dataset.csv')

# Selecting multiple features for PCA
features = ['LotArea', 'OverallQual', 'YearBuilt', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'SalePrice']
X = data[features]

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Creating a PCA model
pca = PCA(n_components=2)  # Reducing the data to 2 dimensions for visualization

# Fitting PCA on the scaled data
X_pca = pca.fit_transform(X_scaled)

# Creating a scatter plot of the first two principal components
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='blue', edgecolor='black', marker='o')
plt.title('PCA of House Pricing Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()

# Explained variance ratio
print("Explained variance ratio of the first two principal components:", pca.explained_variance_ratio_)
```
