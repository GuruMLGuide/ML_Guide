```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()

# Load the dataset
data = pd.read_csv('house price dataset.csv')

# Selecting features for clustering
X = data[['LotArea', 'SalePrice']]  # Using LotArea and SalePrice as features

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Creating a K-Means model
k = 3  # Number of clusters
model = KMeans(n_clusters=k, random_state=42)

# Fitting the model
model.fit(X_scaled)

# Assigning cluster labels to each data point
labels = model.labels_

# Adding cluster labels to the original data
data['Cluster'] = labels

# Plotting the clusters
plt.figure(figsize=(10, 6))
plt.scatter(data['LotArea'], data['SalePrice'], c=data['Cluster'], cmap='viridis', marker='o')
plt.title('Clusters of Houses')
plt.xlabel('Lot Area')
plt.ylabel('Sale Price')
plt.colorbar(label='Cluster')
plt.show()
```
